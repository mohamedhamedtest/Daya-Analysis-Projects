```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"E:\Downloads\udemy_courses_clean.csv")
df
```

```python
df['published_timestamp'] = df['published_timestamp'].str.replace('T', ' ', regex=False)
df['published_timestamp']
```

```python
df['date'] = df['published_timestamp'].str.split(' ').str[0]  # استخراج التاريخ فقط
df['time'] = df['published_timestamp'].str.split(' ').str[1]  # استخراج الوقت فقط
```

```python
df['date']
```

```python
df['time']
```

```python
print(df[['published_timestamp', 'date', 'time']].head())
```

```python
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'], errors='coerce')

df['year'] = df['published_timestamp'].dt.year.fillna(0).astype(int)
df['month'] = df['published_timestamp'].dt.month.fillna(0).astype(int)
df['day'] = df['published_timestamp'].dt.day.fillna(0).astype(int)
```
